#**To remove punctuations and stopwords**


```python
import nltk
from nltk.corpus import stopwords
voc=stopwords.words('english')
import nltk.tokenize as nt
import string

y="It seems like you're referring to Natural Language Processing (NLP). NLP is a field of artificial intelligence that focuses on the interaction between computers and humans through natural language. The goal of NLP is to enable computers to understand, interpret, and generate human language in a way that is both meaningful and useful."
tkz=nt.word_tokenize(y)
filter_words=[i for i in tkz if i not in voc]
pnct=string.punctuation
nopn=[i for i in filter_words if i not in pnct]
print(nopn)
new=""
for i in nopn:
  new+=i+" "
print(new)
```

#**Using Regular expressions**

```python
#word TOKENISATION
import re
tok=re.findall("[\w']+",y)
print(tok)
```

```python
match=re.match("It",y)
print(match.group() if match else "No match")
```

```python
match=re.search("field",y)
print(match.group() if match else "NO result")
```

```python
#sentence TOKENIZE
t1=re.compile("[.?!] ").split(y)
print(t1)
```

```python
pt="[.?!] "
fun=re.compile(pt)
reobj=fun.split(y)
print(reobj)
h="1 2 3 4 5 5"
q=h.split()
print(q)
```

```python
#using finditer()
pt=r"\b\w{3}\b"
fun=re.finditer(pt,y)
for match in fun:
  print("Matched at index: ",match.start())
  print("Matched string: ",match.group())
```

```python
#using sub()
pt=r"\bNLP\b"
fun=re.sub(pt,'nlp',y)
print(fun)
```

```python
#using split()
pt=r"[.?!] "
fun=re.split(pt,y)
print(fun)
```

```python
#using Porterstemmer()
from nltk.stem import PorterStemmer
import re
y="Natural language processing (NLP) is a machine learning technology that gives computers the ability to interpret, manipulate, and comprehend human language. Organizations today have large volumes of voice and text data from various communication channels like emails, text messages, social media newsfeeds, video, audio, and more. They use NLP software to automatically process this data, analyze the intent or sentiment in the message, and respond in real time to human communication."
pt="[\w']+"
fun=re.findall(pt,y)
ps=PorterStemmer()
w=[ps.stem(i) for i in fun]
for j in w:
  print(j)
```

```python
#example for stemming
import nltk.tokenize as nt
from nltk.stem import PorterStemmer
import string
x="The provided code implements the Banker's algorithm for deadlock avoidance in a resource allocation system. Here's a brief explanation of each step:Struct Definition: Defines a structure named da with arrays to hold information about maximum resource needs, allocated resources, resources needed, resources before allocation, and resources after allocation for each process."
c = nt.word_tokenize(x)
pt = PorterStemmer()
wrd =[pt.stem(i) for i in c]
pnc=string.punctuation
y=[i for i in wrd if i not in pnc]
N=""
for i in y:
  N+=i+" "
print(N)

```
